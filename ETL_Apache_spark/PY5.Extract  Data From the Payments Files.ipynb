1.Read Payments File

```python
### DDL format schema
payments_schema = 'payment_id INTEGER,order_id INTEGER,payment_timestamp TIMESTAMP,payment_status INTEGER,payment_method string'
```

```python
df_payments =(
    spark.read.format("csv").
    option("delimeter",",").
    schema(payments_schema).
    load('abfss://gizmobox@deacourseextdl103.dfs.core.windows.net/landing/external_data/payments')
)
```

```python
display(df_payments)
```

```python
df_payments.writeTo("gizmobox.bronze.py_payments").createOrReplace()
```
