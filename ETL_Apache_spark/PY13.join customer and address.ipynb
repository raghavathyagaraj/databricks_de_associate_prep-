### Join Customer Address
Join Customer data with address data to create a customer_address table which conatnis the address of each customer on the same record 

```python
df_customer = spark.table("gizmobox.silver.py_customers")
display(df_customer)
```

```python
df_addresses = spark.table("gizmobox.silver.py_addresses")
display(df_addresses)
```

```python
df_finla =(
    df_customer.
    join(df_addresses, "customer_id","inner").
    select("customer_id",
           "customer_name",
           "date_of_birth",
           "member_since",
           "email",
           "telephone",
           "shipping_address_line_1",
           "shipping_city",
           "shipping_state",
           "shipping_postcode",
           "billing_address_line_1",
           "billing_city",
           "billing_state",
           "billing_postcode")

)
display(df_finla)
```

## write the data to the gold table

```python
df_finla.writeTo("gizmobox.gold.py_customers_addresses").createOrReplace()
```
