### Transform Memberships Data
- 1.Extract customer_id from the file path 
- 2.write transformed data to the silver schema

```python
df_mem = spark.table("gizmobox.bronze.memberships_vw")
```

```python
display(df_mem)
```

1.Extract customer_id from the file path

```python
from pyspark.sql import functions as F
df_final = (
    df_mem.select(
       F.regexp_extract("path",'.*/([0-9]+)\\.png$',1).alias("customer_id"),
       F.col("content").alias("membership_card"))
)
display(df_final)
```

```python
df_final.writeTo("gizmobox.silver.py_membership").createOrReplace()
```
