Reading the addresses file
- 1.Read all the addresses files
- 2.Create addresses table in the bronze layer

1.Read all the addresses files

```python
df_addresses = (spark.read.format("csv")
    .option("header", "true")
    .option("delimiter", "\t")     
    .load("/Volumes/gizmobox/landing/operational_data/addresses"))
```

```python
display(df_addresses)
```

2.Create addresses table in the bronze layer

```python
df_addresses.writeTo("gizmobox.bronze.py_addresses").createOrReplace()
```
