### Monthly order summary 
- 1.total orders
- 2.total items bought
- 3.total amount spent 

```python
df_orders =spark.table("gizmobox.silver.py_order_items")
display(df_orders)
```

```python
from pyspark.sql import functions as F
df_orders_summary =(
    df_orders.withColumn("order_month",F.date_format("transaction_timestamp",'yyyy-MM')).
    groupby("order_month","customer_id").
    agg(F.countDistinct("order_id").alias("total_order"),
        F.sum("quantity").alias("total_items_bought"),
        F.sum(F.col("price") * F.col("quantity")).alias("total_revenue"))
)

display(df_orders_summary)
```

write the data to the gold schema

```python
df_orders_summary.writeTo("gizmobox.gold.py_order_summary_monthly").createOrReplace()
```
