01.Create External Location
- External Location Name :  dea_course_ext_dl_circuitbox
- ADLs Path : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/'
- storage credential : dea-course-ext-sc

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_course_ext_dl_circuitbox
url 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/'
with (STORAGE CREDENTIAL `dea-course-ext-sc`)
COMMENT 'extermal locationf or circuitbox dlt project';
```

```python

%fs ls 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/'
```

### 2.Create Catalog
- catalog_name : circuitbox
- Managed Location : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/'

```python
%sql
CREATE CATALOG IF NOT EXISTS circuitbox
MANAGED LOCATION 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/'
COMMENT 'catalog for the ciruitbox data lakehouse project';
```

```python
%sql
show catalogs;
```

3. Create Schemas 
- 1.**Schema Name:** landing
-     Managed Location : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/landing'
- 2.**Schema Name:** bronze
-     Managed Location : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/lakehouse/bronze'
- 2.**Schema Name:** silver
-     Managed Location : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/lakehouse/silver'
- 2.**Schema Name:** gold
-     Managed Location : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/lakehouse/gold'



```python
%sql
CREATE SCHEMA IF NOT EXISTS circuitbox.landing
MANAGED LOCATION 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/landing/'
COMMENT 'Landing schema for the circuitbox data lakehouse project';

CREATE SCHEMA IF NOT EXISTS circuitbox.bronze
MANAGED LOCATION 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/lakehouse/bronze'
COMMENT 'bronze schema for the circuitbox data lakehouse project';

CREATE SCHEMA IF NOT EXISTS circuitbox.silver
MANAGED LOCATION 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/lakehouse/silver'
COMMENT 'silver schema for the circuitbox data lakehouse project';

CREATE SCHEMA IF NOT EXISTS circuitbox.gold
MANAGED LOCATION 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/lakehouse/gold'
COMMENT 'gold schema for the circuitbox data lakehouse project';
```

**4.Create Volume**
-  Volume Name : operational_data
- ADLS PATH : 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/landing/operational_data'

```python
%sql
CREATE EXTERNAL VOLUME IF NOT EXISTS circuitbox.landing.operational_data
LOCATION 'abfss://circuitbox@deacourseextdl103.dfs.core.windows.net/landing/operational_data';
```
