Create Customer order summary
- 1.join the tables , silver.customers,silver.addresses,silver.orders
- 2.Retrieve the latest address of the customer
- 3.Calculate the following values 
-     total orders
-     total_items_ordered
-     total_orders_amount


```sql
CREATE OR REFRESH MATERIALIZED VIEW circuitbox.gold.customer_order_summary
as 
select  c.customer_id,
        c.customer_name,
        c.date_of_birth,
        c.telephone,
        c.email,
        a.address_line_1,
        a.city,
        a.state,
        a.postcode,
        count(distinct(o.order_id)) as total_orders,
        sum(o.item_quantity) as total_items,
        sum(o.item_quantity * o.item_price) as total_orders_amount
      from circuitbox.silver.customers c
      join circuitbox.silver.addresses a on c.customer_id = a.customer_id
      join circuitbox.silver.orders o on c.customer_id = o.customer_id
      where c.__END_AT is null and a.__END_AT is null 
      group by ALL;
```
