History and Time Travel
- 1.query delta lake table history
- 2.query previous versions of the data 
- 3.query data from a specific time
- 4.Restore data to a specific verison

1.query delta lake table history

```python
%sql
describe history demo.delta_lake.companies
```

```python
%sql
select * from demo.delta_lake.companies version as of 1
```

```python
%sql
select * from demo.delta_lake.companies version as of 0;
```

```python
%sql
select * from demo.delta_lake.companies timestamp as of '2025-12-24T19:14:41.000+00:00'
```

4.Restore the data in the table to specific version

```python
%sql
RESTORE TABLE demo.delta_lake.companies VERSION AS OF 1;
```

```python
%sql
select * from demo.delta_lake.companies;
```

```python
%sql
describe history demo.delta_lake.companies;
```
