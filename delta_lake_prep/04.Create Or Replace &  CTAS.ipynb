1.Difference between Create or Replace and Drop and Create Table Statements

Behaviour of DROP and CREATE statements

```python
%sql
drop table if exists demo.delta_lake.companies;
CREATE  table demo.delta_lake.companies
(
  company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  company_name string,
  founding_date DATE comment 'this columns represents the founding date of the company',
  country string,
  founded_year INT GENERATED ALWAYS AS (YEAR(founding_date))
)
comment 'This is a table of companies'
TBLPROPERTIES("sensitive"="true","delta.enableDeletionVectors"= "false")
```

```python
%sql
INSERT INTO demo.delta_lake.companies(company_name, founding_date, country)
values ("apple", "2007-01-01", "USA");
```

```python
%sql
describe history demo.delta_lake.companies;
```

Behaviour of DROP and CREATE statements

### THIS IS THE RECOMMEND METHOD RATHER THAN DROP AND CREATE 

```python
%sql
drop table if exists demo.delta_lake.companies;
```

```python
%sql
CREATE OR REPLACE table demo.delta_lake.companies
(
  company_id BIGINT NOT NULL GENERATED ALWAYS AS IDENTITY (START WITH 1 INCREMENT BY 1),
  company_name string,
  founding_date DATE comment 'this columns represents the founding date of the company',
  country string,
  founded_year INT GENERATED ALWAYS AS (YEAR(founding_date))
)
comment 'This is a table of companies'
TBLPROPERTIES("sensitive"="true","delta.enableDeletionVectors"= "false")
```

```python
%sql
INSERT INTO demo.delta_lake.companies(company_name, founding_date, country)
values ("apple", "2007-01-01", "USA");
```

```python
%sql
INSERT INTO demo.delta_lake.companies(company_name, founding_date, country)
values ("samsung", "2107-01-01", "china");
```

```python
%sql
describe history demo.delta_lake.companies;
```

2.CTAS statement

```python
%sql
select * from demo.delta_lake.companies where country = 'china';
```

```python
%sql
CREATE TABLE demo.delta_lake.companies_china 
as 
select * from demo.delta_lake.companies
 where country = 'china';
```

```python
%sql
DESCRIBE HISTORY demo.delta_lake.companies_china;
```

Limitations of the CTAS statements :
- You cannot specify the column properties 

```python
%sql
desc demo.delta_lake.companies_china;
```

```python
%sql
drop table demo.delta_lake.companies_china ;
```

```python
%sql
CREATE TABLE demo.delta_lake.companies_china 
as 
select cast(company_id as int) as company_id,
        company_name,
        founding_date,
        country,
        founded_year
         from demo.delta_lake.companies
 where country = 'china';
```

```python
%sql
select * from demo.delta_lake.companies_china ;
```

But we can add the comments for the column using the alter statement 

```python
%sql
alter table demo.delta_lake.companies_china
  alter column founding_date comment 'This is Company Founded date';
```

```python
%sql
desc demo.delta_lake.companies_china;
```

```python
%sql
alter table demo.delta_lake.companies_china
  alter column company_id set not null;
```
