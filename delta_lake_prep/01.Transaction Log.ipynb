#### understaing Delta Lake Transaction Log
- understanding the cloud storage directory structure behind delta lake tables

- Create a new schema under the demo catalog for this section of the course(delta_lake)

```python
%sql
CREATE CATALOG IF NOT EXISTS demo
MANAGED LOCATION 'abfss://demo@deacourseextdl103.dfs.core.windows.net/'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
MANAGED LOCATION 'abfss://demo@deacourseextdl103.dfs.core.windows.net/delta_lake'
```

1.Create Delta Lake Table

```python
%sql
CREATE TABLE IF NOT EXISTS demo.delta_lake.companies
(
  company_name STRING,
  founded_date DATE,
  country STRING
)
```

```python
%sql
DESCRIBE EXTENDED demo.delta_lake.companies;
```
